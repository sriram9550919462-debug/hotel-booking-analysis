```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('hotel_booking.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df['hotel'].value_counts()
```

```python
df['hotel'].value_counts().plot(kind='bar')
plt.title('book by hotel type')
plt.xlabel('hotel type ')
plt.ylabel('number of boings')
```

```python
df['is_canceled'].value_counts()
```

```python
df['is_canceled'].value_counts().plot(kind='bar')
plt.title('canceled vs not canceled bookings')
plt.xlabel('0= not canceled, 1 = canceled' )
plt.ylabel('number of bookings')

```

```python
df.groupby('hotel')['is_canceled'].mean()
```

```python
df.groupby('hotel')['is_canceled'].mean().plot(kind="bar")
plt.title('cancellation rate by hotel types')
plt.xlabel('hotel types')
plt.ylabel('cancellation rate')
plt.show()
```

```python
df['arrival_date_month'].value_counts()
```

```python
months=[

    'January','February','March','April','May','June',
    'july','August','October','September','November','December',
]
df['arrival_date_month'] = pd.Categorical(
    df['arrival_date_month'],
    categories=months,
    ordered=True
)
df['arrival_date_month'].value_counts().sort_index().plot(kind='line')
plt.title("Monthly Booking Trend")
plt.xlabel("Month")
plt.ylabel("Bookings")
plt.show()
```

```python
df.groupby('hotel')['adr'].mean()
```

```python
df.groupby('hotel')['adr'].mean().plot(kind='bar')
plt.title('Average Daily rate by hotel type')
plt.xlabel('hotel type')
plt.ylabel('ADR')
```
